```python
## Train Val split

import os
import pandas as pd

label_path = "./dataset/label.csv"
train_path = "./dataset/train_label.csv"
val_path = "./dataset/val_label.csv"
val_num = 2


df=pd.read_csv(label_path)

train_list = []
val_list = []
all_list = []
cnt = 0

for index, row in df.iterrows():
    # print(row["filename"], row["category"])
    if row["category"] == cnt:
        cnt = cnt
    else: 
        cnt = (cnt + 1)
        for i in range(val_num):
            val_list.append(all_list[i])
        for i in range(val_num, len(all_list)):
            train_list.append(all_list[i])
        all_list = []

    all_list.append([row["filename"], cnt])

for i in range(val_num):
    val_list.append(all_list[i])
for i in range(val_num, len(all_list)):
    train_list.append(all_list[i])

train_df = pd.DataFrame(train_list, columns=["filename", "category"])
val_df = pd.DataFrame(val_list, columns=["filename", "category"])

train_df.to_csv(train_path, index=False)
val_df.to_csv(val_path, index=False)
```
